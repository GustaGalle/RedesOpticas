# Análise de Rede Óptica Completa

## 1. Parâmetros da Rede

- **Número de nós (N):** 20  
  Todos os nós interconectados (topologia completa).

- **Dimensões físicas:**  
  - Largura: $150\,\text{km}$  
  - Altura: $150\,\text{km}$  
  - Área total: $A = 150 \times 150 = 22,\!500\,\text{km}^2$

---

## 2. Conectividade da Rede

### Grau Médio ($\langle k \rangle$)
$$
\langle k \rangle = \frac{\text{Total de conexões}}{N} = \frac{190}{20} = 19
$$

### Grau Máximo ($k_{\text{max}}$)
$$
k_{\text{max}} = N - 1 = 19
$$

### Número Total de Conexões
$$
E = \frac{N(N-1)}{2} = \frac{20 \times 19}{2} = 190\,\text{conexões}
$$

---

## 3. Análise Geométrica

### Distância Máxima (Diagonal)
$$
d_{\text{max}} = \sqrt{\text{Largura}^2 + \text{Altura}^2} = \sqrt{150^2 + 150^2} \approx 212.13\,\text{km}
$$

### Área de Cobertura
$$
A_{\text{rede}} = \pi \left(\frac{d_{\text{max}}}{2}\right)^2 \approx 35,\!343\,\text{km}^2
$$

##  **Parâmetros do Transceptor**

### 1.1. **Select Transceiver Type**
- **Significado**: Define o tipo de transceptor usado na rede.
- **Exemplos**:
  - Transceptores de alta capacidade (ex.: 400G, 800G).
  - Transceptores de baixa capacidade (ex.: 100G).
- **Impacto**:
  - O tipo de transceptor afeta diretamente a taxa de dados, a sensibilidade e os requisitos de qualidade do sinal (SNR/OSNR).

### 1.2. **Select Transceiver Mode**
- **Significado**: Define o modo de operação do transceptor.
- **Exemplos**:
  - Modulação (ex.: QPSK, 16-QAM, 64-QAM).
  - Taxa de bits (ex.: 100G, 200G, 400G).
- **Impacto**:
  - Modulações mais complexas (como 64-QAM) permitem maior capacidade, mas exigem maior SNR/OSNR.
  - Modulações simples (como QPSK) são mais robustas, mas têm menor capacidade.

### 1.3. **Margin [dB]**
- **Significado**: Margem de segurança para garantir que o sistema funcione corretamente, mesmo em condições adversas.
- **Default**: `0 dB`
- **Impacto**:
  - Um valor positivo aumenta a margem de segurança, mas pode reduzir a eficiência da rede.
  - Exemplo: Se o SNR mínimo for 20 dB e a margem for 3 dB, o sistema será projetado para atingir pelo menos 23 dB.

### 1.4. **Baudrate [Gbaud]**
- **Significado**: Taxa de símbolos transmitidos por segundo.
- **Default**: `31.57 Gbaud`
- **Impacto**:
  - A taxa de baud determina a largura de banda ocupada pelo sinal.
  - Exemplo: Um baudrate de 31.57 Gbaud é típico para sistemas de 100G ou 200G.

### 1.5. **Roll-off**
- **Significado**: Fator de roll-off do filtro de formatação de pulsos.
- **Default**: `0.15`
- **Impacto**:
  - Controla a largura espectral adicional necessária para evitar interferências entre símbolos.
  - Um valor menor reduz a largura de banda, mas pode aumentar a interferência.

### 1.6. **Tx OSNR [dB]**
- **Significado**: Relação Sinal-Ruído Óptico (OSNR) do sinal transmitido.
- **Default**: `35 dB`
- **Impacto**:
  - Um valor alto indica um sinal limpo e de alta qualidade.
  - Afeta diretamente a qualidade do sinal ao longo da rede.

### 1.7. **Min OSNR [dB]**
- **Significado**: Valor mínimo de OSNR necessário para que o receptor funcione corretamente.
- **Impacto**:
  - Define o limite de qualidade do sinal na recepção.
  - Valores típicos dependem do tipo de modulação (ex.: 18 dB para QPSK, 25 dB para 16-QAM).

---

## 2. **Seleção de Nós de Origem e Destino**

### 2.1. **Select Source Node / Destination Node**
- **Significado**: Define os nós de origem e destino para o tráfego óptico.
- **Impacto**:
  - Determina o caminho que o sinal seguirá na rede.
  - Importante para calcular o desempenho do sinal ao longo do caminho.

### 2.2. **Select Nodes to Include / Exclude**
- **Significado**: Permite incluir ou excluir nós específicos no caminho.
- **Impacto**:
  - Útil para testar cenários específicos ou evitar nós problemáticos.

### 2.3. **Preserve Order**
- **Significado**: Garante que os nós sejam percorridos na ordem especificada.
- **Impacto**:
  - Importante para redes onde a ordem dos nós afeta o desempenho (ex.: amplificadores intermediários).

---

## 3. **Referência de Potência e Varredura**

### 3.1. **Reference Power**
- **Significado**: Potência de referência usada para simular o sinal óptico.
- **Default**: `2.0 dBm`
- **Impacto**:
  - Define o nível de potência inicial do sinal.
  - Valores muito baixos podem causar atenuação excessiva.

### 3.2. **Sweep Parameters (Start, Stop, Step)**
- **Significado**: Realiza uma varredura de potência para testar diferentes níveis de transmissão.
- **Default**:
  - Start: `2.0 dBm`
  - Stop: (opcional)
  - Step: (opcional)
- **Impacto**:
  - Útil para encontrar o nível ideal de potência para maximizar o desempenho.

---

## 4. **Parâmetros Espectrais**

### 4.1. **Spacing [GHz]**
- **Significado**: Espaçamento entre canais no domínio da frequência.
- **Default**: `50 GHz`
- **Impacto**:
  - Define a densidade espectral da rede.
  - Espaçamentos menores (ex.: 37.5 GHz) permitem mais canais, mas aumentam a interferência.

### 4.2. **Minimum Frequency [THz]**
- **Significado**: Frequência mínima do espectro óptico.
- **Default**: `191.30 THz`
- **Impacto**:
  - Define o limite inferior do espectro disponível.

### 4.3. **Maximum Frequency [THz]**
- **Significado**: Frequência máxima do espectro óptico.
- **Default**: `196.10 THz`
- **Impacto**:
  - Define o limite superior do espectro disponível.

---

## 5. **Parâmetros do ROADM**

### 5.1. **Add/Drop OSNR [dB]**
- **Significado**: Relação Sinal-Ruído Óptico (OSNR) nos portos de adição/remoção de canais (ROADM).
- **Default**: `30 dB`
- **Impacto**:
  - Afeta a qualidade do sinal ao adicionar/remover canais em nós intermediários.

## Adicionando Parâmetros Físicos

### Atenuação Óptica
$$
\alpha = 0.2\,\text{dB/km} \\
P_{\text{perda}} = \alpha \times d_{\text{max}} \approx 42.43\,\text{dB}
$$


### Latência de Propagação
$$
t_{\text{latência}} = \frac{d_{\text{max}}}{200,\!000\,\text{km/s}} \approx 1.06\,\text{ms}
$$



```python
import networkx as nx
import matplotlib.pyplot as plt
import math

# Parâmetros da rede
num_nos = 20
largura = 150  # km
altura = 150   # km

# Criar grafo completo (todos conectados a todos)
G = nx.complete_graph(num_nos)

#  Calcular propriedades
grau_medio = sum(dict(G.degree()).values()) / num_nos
grau_max = max(dict(G.degree()).values())
num_arestas = G.number_of_edges()

# Posicionar nós aleatoriamente na área de 150x150 km
import random
posicoes = {no: (random.uniform(0, largura), random.uniform(0, altura)) for no in G.nodes()}

# Calcular distâncias e atenuação
distancias = []
for u, v in G.edges():
    x1, y1 = posicoes[u]
    x2, y2 = posicoes[v]
    distancia = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    distancias.append(distancia)

distancia_maxima = max(distancias)
atenuacao_maxima = 0.2 * distancia_maxima  # 0.2 dB/km


plt.figure(figsize=(10, 8))
nx.draw(G, pos=posicoes, node_size=300, node_color='lightblue', with_labels=True)
plt.title(f"Rede Óptica Completa (20 nós, 190 conexões)\nDistância máxima: {distancia_maxima:.2f} km")
plt.xlabel("Largura (km)")
plt.ylabel("Altura (km)")
plt.grid(True)
plt.show()

# resultados
print("\n--- Análise da Rede ---")
print(f"Número de nós: {num_nos}")
print(f"Grau médio: {grau_medio} (teórico: 19)")
print(f"Grau máximo: {grau_max} (teórico: 19)")
print(f"Número de arestas (conexões): {num_arestas} (teórico: 190)")
print(f"Distância máxima entre nós: {distancia_maxima:.2f} km (teórico: ~212 km)")
print(f"Atenuação máxima (0.2 dB/km): {atenuacao_maxima:.2f} dB")
print(f"Latência máxima (200.000 km/s): {distancia_maxima / 200000:.4f} segundos")

```
